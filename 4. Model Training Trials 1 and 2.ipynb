```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
import joblib

# Load your DataFrame (assuming the CSV file is already saved)
dataset = pd.read_csv("COMPLETED_CLASSIFIED_DATASET.csv")

# Task 1: Binary Classification
print("Preparing data for binary classification...")
X_binary = dataset['text']
y_binary = dataset['label']

# Balance the dataset
ros_binary = RandomOverSampler(random_state=42)
X_binary_balanced, y_binary_balanced = ros_binary.fit_resample(X_binary.to_frame(), y_binary)

# TF-IDF Vectorization for binary classification
print("Generating TF-IDF features for binary classification...")
tfidf_vectorizer_binary = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf_binary = tfidf_vectorizer_binary.fit_transform(X_binary_balanced['text'])

# Split data for binary classification
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_tfidf_binary, y_binary_balanced, test_size=0.2, random_state=42
)

# Train Random Forest Classifier for binary classification
print("Training Random Forest classifier for binary classification...")
rf_model_binary = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf_model_binary.fit(X_train_b, y_train_b)

# Evaluate the binary classification model
y_binary_pred = rf_model_binary.predict(X_test_b)
binary_class_accuracy = accuracy_score(y_test_b, y_binary_pred)

print("Task 1 - Binary Classification Accuracy:", binary_class_accuracy)
print("Task 1 - Classification Report:\n", classification_report(y_test_b, y_binary_pred))

# Save the binary classification model and vectorizer
joblib.dump(rf_model_binary, 'binary_random_forest_classifier.pkl')
joblib.dump(tfidf_vectorizer_binary, 'tfidf_vectorizer_binary.pkl')

print("Binary classification model and vectorizer saved successfully.")

# Task 2: Multi-Class Classification
legitimate_data = dataset[dataset['label'] == 1]
X_classification = legitimate_data['text']
y_classification = legitimate_data['class']

# Balance the dataset
ros = RandomOverSampler(random_state=42)
X_classification_balanced, y_classification_balanced = ros.fit_resample(X_classification.to_frame(), y_classification)

# Encode class labels as numeric values
label_encoder = LabelEncoder()
y_classification_encoded = label_encoder.fit_transform(y_classification_balanced)

# TF-IDF Vectorization
print("Generating TF-IDF features...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = tfidf_vectorizer.fit_transform(X_classification_balanced['text'])

# Split data for multi-class classification
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_tfidf, y_classification_encoded, test_size=0.2, random_state=42
)

# Train Random Forest Classifier
print("Training Random Forest classifier...")
rf_model = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf_model.fit(X_train_c, y_train_c)

# Evaluate the model
y_multi_pred = rf_model.predict(X_test_c)
multi_class_accuracy = accuracy_score(y_test_c, y_multi_pred)

# Decode the predicted labels back to original string values for reporting
y_test_decoded = label_encoder.inverse_transform(y_test_c)
y_pred_decoded = label_encoder.inverse_transform(y_multi_pred)

print("Task 2 - Multi-Class Classification Accuracy:", multi_class_accuracy)
print("Task 2 - Classification Report:\n", classification_report(y_test_decoded, y_pred_decoded))

# Save the model, vectorizer, and label encoder
joblib.dump(rf_model, 'random_forest_classifier.pkl')
joblib.dump(tfidf_vectorizer, 'tfidf_vectorizer.pkl')
joblib.dump(label_encoder, 'label_encoder.pkl')

print("Model, vectorizer, and label encoder saved successfully.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
import joblib
from xgboost import XGBClassifier

# Load your DataFrame (assuming the CSV file is already saved)
dataset = pd.read_csv("COMPLETED_CLASSIFIED_DATASET.csv")

# Task 1: Binary Classification
print("Preparing data for binary classification...")
X_binary = dataset['text']
y_binary = dataset['label']

# Balance the dataset
ros_binary = RandomOverSampler(random_state=42)
X_binary_balanced, y_binary_balanced = ros_binary.fit_resample(X_binary.to_frame(), y_binary)

# TF-IDF Vectorization for binary classification
print("Generating TF-IDF features for binary classification...")
tfidf_vectorizer_binary = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf_binary = tfidf_vectorizer_binary.fit_transform(X_binary_balanced['text'])

# Split data for binary classification
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_tfidf_binary, y_binary_balanced, test_size=0.2, random_state=42
)

# Train XGBoost Classifier for binary classification
print("Training XGBoost classifier for binary classification...")
xgb_model_binary = XGBClassifier(use_label_encoder=False, eval_metric='logloss', n_estimators=200, max_depth=10, learning_rate=0.1)
xgb_model_binary.fit(X_train_b, y_train_b)

# Evaluate the binary classification model
y_binary_pred = xgb_model_binary.predict(X_test_b)
binary_class_accuracy = accuracy_score(y_test_b, y_binary_pred)

print("Task 1 - Binary Classification Accuracy:", binary_class_accuracy)
print("Task 1 - Classification Report:\n", classification_report(y_test_b, y_binary_pred))

# Save the binary classification model and vectorizer
joblib.dump(xgb_model_binary, 'binary_xgb_classifier.pkl')
joblib.dump(tfidf_vectorizer_binary, 'tfidf_vectorizer_binary.pkl')

print("Binary classification model and vectorizer saved successfully.")

# Task 2: Multi-Class Classification
legitimate_data = dataset[dataset['label'] == 1]
X_classification = legitimate_data['text']
y_classification = legitimate_data['class']

# Balance the dataset
ros = RandomOverSampler(random_state=42)
X_classification_balanced, y_classification_balanced = ros.fit_resample(X_classification.to_frame(), y_classification)

# Encode class labels as numeric values
label_encoder = LabelEncoder()
y_classification_encoded = label_encoder.fit_transform(y_classification_balanced)

# TF-IDF Vectorization
print("Generating TF-IDF features...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = tfidf_vectorizer.fit_transform(X_classification_balanced['text'])

# Split data for multi-class classification
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_tfidf, y_classification_encoded, test_size=0.2, random_state=42
)

# Train XGBoost Classifier
print("Training XGBoost classifier...")
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', n_estimators=200, max_depth=10, learning_rate=0.1)
xgb_model.fit(X_train_c, y_train_c)

# Evaluate the model
y_multi_pred = xgb_model.predict(X_test_c)
multi_class_accuracy = accuracy_score(y_test_c, y_multi_pred)

# Decode the predicted labels back to original string values for reporting
y_test_decoded = label_encoder.inverse_transform(y_test_c)
y_pred_decoded = label_encoder.inverse_transform(y_multi_pred)

print("Task 2 - Multi-Class Classification Accuracy:", multi_class_accuracy)
print("Task 2 - Classification Report:\n", classification_report(y_test_decoded, y_pred_decoded))

# Save the model, vectorizer, and label encoder
joblib.dump(xgb_model, 'multi_class_classifier.pkl')
joblib.dump(tfidf_vectorizer, 'tfidf_vectorizer.pkl')
joblib.dump(label_encoder, 'label_encoder.pkl')

print("Model, vectorizer, and label encoder saved successfully.")
```
